```python
%%R
library(data.table)
library(tidyverse)
library(soiltexture)
library(ggtern)
library(compositions)
library(robCompositions)
library(FactoMineR)
library(factoextra)
library(plyr)
library(grid)
```

```python
%%R
#Read the data file
soildata <- read.csv('data.csv')
str(soildata)
summary(soildata)
```

```python
%%R
soil_df <- data.frame(
    "CLAY" = soildata$claytotal_r,
    "SILT" = soildata$silttotal_r,
    "SAND" = soildata$sandtotal_r,
    "field_name"=soildata$field_name,
    "sample_uuid"=soildata$sample_uuid
)
```

```python
%%R
#visuals for soil category
TT.plot( 
  class.sys   = "USDA-NCSS.TT", 
  class.p.bg.col  = TRUE,
  tri.data    = soil_df, 
  main        = "Soil texture data" 
) 
```

```python
%%R
#visual for soil content
soil_df %>%
  ggtern(aes(x=SAND, y=CLAY, z=SILT)) + 
  stat_density_tern(aes(fill = ..level.., alpha = ..level..), geom = "polygon") +
  scale_fill_gradient2(low="red", high="blue") +
  theme_bw(base_size=14) +
  guides(color="none", fill="none", alpha="none")
```

```python
%%R
#code to classify soil texture
soil_df <- cbind(soil_df, 
                "TEXCLASS" = TT.points.in.classes(tri.data  = soil_df[, c('CLAY', 'SILT',  'SAND')],
                                                     css.names = c('CLAY', 'SILT', 'SAND'),
                                                     class.sys = "USDA-NCSS.TT", 
                                                     PiC.type  = "t",
                                                     collapse  = ', ')
                  )

str(soil_df)
```

```python
%%R
write.csv(x=soil_df, file="soiltxt.csv")
```
